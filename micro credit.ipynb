```python
def metric_score(clf,x_train,x_test,y_train,y_test,train=True):
    if train:
        y_pred=clf.predict(x_train)
        print("=======train result==========")
        print(f"accuracy score:{accuracy_score(y_train,y_pred)*100:.2f}%")
    elif train==False:
        pred=clf.predict(x_test)
        print("=======test result==========")
        print(f"accuracy score:{accuracy_score(y_test,pred)*100:.2f}%")
        
```

```python
#score with k neighbors classifier
```

```python
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
```

```python
#calling the function and checking the score
```

```python
metric_score(knn,x_train,x_test,y_train,y_test,train=True)
metric_score(knn,x_train,x_test,y_train,y_test,train=False)
```

```python
# hyper parameter tuning
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'algorithm':['kd-tree','brute'],
           'leaf_size':[3,5,6,7,8],
           'n_neighbors':[3,5,7,9,11,13]}
grd=GridSearchCV(estimator=knn,param_grid=param_grid)
grd.fit(x_train,y_train)
grd.best_params_
```

```python
knn=KNeighborsClassifier(algorithm='brute',leaf_size=3,n_neighbors=13)
knn.fit(x_train,y_train)
```

```python
metric_score(knn,x_train,x_test,y_train,y_test,train=True)
metric_score(knn,x_train,x_test,y_train,y_test,train=False)
```

```python
clf=DecisionTreeClassifier()
clf.fit(x_train,y_train)
```

```python
metric_score(clf,x_train,x_test,y_train,y_test,train=True)
metric_score(clf,x_train,x_test,y_train,y_test,train=False)
```

```python
grid_param={'criterion':['gini','entropy'],
            'max_depth':range(10,15),
            'min_samples_leaf':range(2,6),
            'min_samples_split':range(3,8),
            'max_leaf_nodes':range(5,10)}
grd_srch=GridSearchCV(estimator=clf,param_grid=grid_param,n_jobs=-1)
grd_srch.fit(x_train,y_train)
    
```

```python
best_parameters=grd_srch.best_params_
print(best_parameters)
```

```python
clf=DecisionTreeClassifier(criterion='entropy',max_depth=10,max_leaf_nodes=8,min_samples_leaf=2,min_samples_split=3)
clf.fit(x_train,y_train)
```
